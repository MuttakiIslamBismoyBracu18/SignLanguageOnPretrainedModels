```python
import os
import shutil
import random
```

```python
# Define the base directory where the dataset is stored
base_dir = r"E:\Undergrad Research\DATASET_MIXED\Train"
test_dir = r"E:\Undergrad Research\DATASET_MIXED\Test"
```

```python
# Ensure the test directory exists
os.makedirs(test_dir, exist_ok=True)
```

```python
# Function to create a 10% test set
def create_test_set(base_dir, test_dir, test_size=0.1):
    for subdir in os.listdir(base_dir):
        subdir_path = os.path.join(base_dir, subdir)
        if os.path.isdir(subdir_path):
            # Create corresponding sub-directory in the test directory
            test_subdir_path = os.path.join(test_dir, subdir)
            os.makedirs(test_subdir_path, exist_ok=True)
            
            # Get list of all images in the sub-directory
            images = [f for f in os.listdir(subdir_path) if f.endswith(('jpg', 'jpeg', 'png'))]
            
            # Calculate the number of images to move to the test set
            num_test_images = int(len(images) * test_size)
            
            # Randomly select images to move to the test set
            test_images = random.sample(images, num_test_images)
            
            # Move the selected images to the test set
            for img in test_images:
                src_path = os.path.join(subdir_path, img)
                dst_path = os.path.join(test_subdir_path, img)
                shutil.move(src_path, dst_path)
                print(f"Moved {src_path} to {dst_path}")
```

```python
# Run the function to create the test set
create_test_set(base_dir, test_dir)
```
